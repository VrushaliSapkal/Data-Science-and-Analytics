## **Adding a column to the dataset** 

If want to add a new column in our dataset, as we are doing any calculations or extracting some information from the dataset, and if you want to save it a new column.

```python
# For example, if we want to add a column let say "total_values",
# that means if you want to add all the integer value of that particular
# row and get total answer in the new column "total_values".
# first we will extract the columns which have integer values.
cols = df.columns

# it will print the list of column names.
print(cols)

# we will take that columns which have integer values.
cols = cols[1:5]

# we will save it in the new dataframe variable
data1 = df[cols]

# now adding new column "total_values" to dataframe data.
df["total_values"]=data1[cols].sum(axis=1)

# here axis=1 means you are working in rows,
# whereas axis=0 means you are working in columns.
```

**Heatmap**:

1. The heatmap is a data visualisation technique which is used to analyse the dataset as colors in two dimensions. 
2. Basically it shows correlation between all numerical variables in the dataset. 
3. Heatmap is an attribute of the Seaborn library.
4. Annotate each cell with the numeric value using integer formatting
 


```python
corr = df.corr()
fig , ax = plt.subplots(figsize=(10,5))
sns.heatmap(corr , annot=True, ax=ax, cmap = 'coolwarm',linecolor = 'white',linewidths = 1)
```

```python
corr = df.corr()
fig , ax = plt.subplots(figsize=(10,5))
sns.heatmap(corr , annot=True, ax=ax)
```

**Label Encoder**

1. In machine learning we usually deal with datasets which contains multiple labels in one or more than one columns.
2. These labels can be in the form of words or numbers.
3. Label encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form.


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
df.head()
```
